```python
# Create figure and axes
fig, ax = plt.subplots(figsize=(8, 10))

# Title and headers
ax.text(0.5, 1.05, "CORE VOLLEYBALL LINEUP SHEET", fontsize=14, ha='center', va='center', transform=ax.transAxes)
ax.text(0.1, 1.01, "Team Name: ____________________", fontsize=10, ha='left', va='center', transform=ax.transAxes)
ax.text(0.6, 1.01, "AGE: ____   Date: ___________", fontsize=10, ha='left', va='center', transform=ax.transAxes)

# Draw the player list table
ax.plot([0.1, 0.4, 0.4, 0.1, 0.1], [0.9, 0.9, 0.4, 0.4, 0.9], 'k')  # Table outline
ax.text(0.2, 0.98, "PLAYER NAMES", fontsize=10, ha='center', va='center', transform=ax.transAxes)

# Add rows for players
for i in range(10):
    y = 0.9 - (i + 1) * 0.045
    ax.plot([0.1, 0.4], [y, y], 'k')

# Draw lineup boxes (three sets of six positions)
x_start = 0.5
y_start = 0.9
box_height = 0.15
for set_index in range(2):
    # Draw the box for the set
    box_y = y_start - set_index * (box_height + 0.05)
    ax.plot([x_start, x_start + 0.35, x_start + 0.35, x_start, x_start],
            [box_y, box_y, box_y - box_height, box_y - box_height, box_y], 'k')
    
    # Divide the box into 6 positions
    ax.plot([x_start + 0.175, x_start + 0.175], [box_y, box_y - box_height], 'k')  # Vertical line
    ax.plot([x_start, x_start + 0.35], [box_y - box_height / 2, box_y - box_height / 2], 'k')  # Horizontal line


# Add footnote
ax.text(0.98, 0.2, "Place the uniform number of each player in the box according to their starting position on the court.", 
        fontsize=8, ha='center', transform=ax.transAxes)
ax.text(0.98, 0.1, "Server will begin in Position 1 if team is serving first and in Position 2 if serving second.", 
        fontsize=9, ha='center', transform=ax.transAxes)

# Hide axes
ax.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create figure and axes
fig, ax = plt.subplots(figsize=(8, 10))

# Title and headers
ax.text(0.5, 1.05, "CORE VOLLEYBALL LINEUP SHEET", fontsize=14, ha='center', va='center', transform=ax.transAxes)
ax.text(0.1, 1.01, "Team Name: _________________", fontsize=10, ha='left', va='center', transform=ax.transAxes)
ax.text(0.6, 1.01, "AGE: ___   Date: ___________", fontsize=10, ha='left', va='center', transform=ax.transAxes)

# Draw the player list table
ax.plot([0.1, 0.4, 0.4, 0.1, 0.1], [0.9, 0.9, 0.4, 0.4, 0.9], 'k')  # Table outline
ax.text(0.2, 0.98, "PLAYER NAMES", fontsize=10, ha='center', va='center', transform=ax.transAxes)

# List of player names
player_names = [
    "Anessa", "Isabell", "Niat", "Olivia", "Julianna",
    "Alanna", "Esther", "Avery", "Claire", "Layla"
]

# Add rows for players and populate names
row_height = 0.05
for i, name in enumerate(player_names):
    y_top = 0.9 - (i * row_height)  # Top of the row
    y_center = y_top - row_height / 2  # Center of the row
    ax.plot([0.1, 0.4], [y_top, y_top], 'k')  # Horizontal row line

# Draw court with 6 positions
court_x_start = 0.5
court_y_start = 0.9
court_width = 0.35
court_height = 0.15

for set_index in range(3):
    # Top-left corner of the court for this set
    box_y = court_y_start - set_index * (court_height + 0.05)
    
    # Draw outer box
    ax.plot([court_x_start, court_x_start + court_width, court_x_start + court_width, court_x_start, court_x_start],
            [box_y, box_y, box_y - court_height, box_y - court_height, box_y], 'k')
    
    # Divide into 6 positions
    row_height = court_height / 3
    col_width = court_width / 2
    
    # Draw horizontal lines
    for i in range(1, 3):
        ax.plot([court_x_start, court_x_start + court_width], [box_y - i * row_height, box_y - i * row_height], 'k')
    
    # Draw vertical line
    ax.plot([court_x_start + col_width, court_x_start + col_width], [box_y, box_y - court_height], 'k')
    


# Hide axes
ax.axis('off')

# Show the plot
plt.tight_layout()
# Save the plot with high DPI for printing quality
plt.savefig('graph.pdf', dpi=300)

# Show the plot
plt.show()
```
